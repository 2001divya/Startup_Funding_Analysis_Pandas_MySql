```python
from sqlalchemy import create_engine
import pandas as pd
from sqlalchemy.types import (SmallInteger, Date, String, Float, Text)
```

```python
engine = create_engine("mysql+pymysql://root:123456@localhost/startup_funding")
```

```python
cols = {
    "sr_no": SmallInteger(),
    "funding_date": Date(),
    "startup_name": String(75),
    "industry_vertical": String(100),
    "sub_vertical": String(150),
    "city_location": String(50),
    "investors_name": String(255),
    "investment_type": String(50),
    "amount_usd": Float(),
    "remarks": Text()
}
df=pd.read_csv(r'D:\Programming\Python\Data Analysis\Startup_Funding_Analysis\data\startup_clean.csv')
df.to_sql("startup_clean",con=engine,if_exists="replace",index=False,dtype=cols)
```
